```python
q1_am = (df['Q1'] > df['Q1'].mean()).sum()
q1_bm = (~(df['Q1'] > df['Q1'].mean())).sum()
q2_am = (df['Q2'] > df['Q2'].mean()).sum()
q2_bm = (~(df['Q2'] > df['Q2'].mean())).sum()
q3_am = (df['Q3'] > df['Q3'].mean()).sum()
q3_bm = (~(df['Q3'] > df['Q3'].mean())).sum()
```

```python
am = [q1_am, q2_am, q3_am]
bm = [q1_bm, q2_bm, q3_bm ]
```

```python
def autolabel1(rects, i):
    for j, rect in enumerate(rects):
        height = rect.get_height()
        
        if i == 0:
            ax.text(rect.get_x() - 0.1, 0.8*height/2.,
               '%d' % int(height),
                ha='center', va='bottom')
            h1 = height
        else:
            ax.text(rect.get_x() - 0.1, 0.8*height/2. + bm[j],
               '%d' % int(height),
                ha='center', va='bottom')
```

```python
ax = plt.subplot(111)
q1_am = (df['Q1'] > df['Q1'].mean()).sum()
q1_bm = (~(df['Q1'] > df['Q1'].mean())).sum()
q2_am = (df['Q2'] > df['Q2'].mean()).sum()
q2_bm = (~(df['Q2'] > df['Q2'].mean())).sum()
q3_am = (df['Q3'] > df['Q3'].mean()).sum()
q3_bm = (~(df['Q3'] > df['Q3'].mean())).sum()

ind = np.arange(3)
am = [q1_am, q2_am, q3_am]
bm = [q1_bm, q2_bm, q3_bm ]

autolabel1(ax.bar(ind, bm,  width=0.15, color='r', align='center'),0)
autolabel1(ax.bar(ind, am,  width=0.15, color='b', bottom=bm, align='center'),1)

plt.xticks(ind, ['CO1' + '\nMean = %.2f' % (df['Q1'].mean()/2.0), 
                 'CO2' + '\nMean = %.2f' % df['Q2'].mean(),
                 'CO3' + '\nMean = %.2f' % df['Q3'].mean()])
plt.legend(['Below Mean', 'Above Mean'], loc='best', bbox_to_anchor=(1, 1))
plt.title('Performance Analysis')
plt.show()

print('Above Mean:' ,q1_am, q2_am, q3_am)
print('Below Mean:' ,q1_bm, q2_bm, q3_bm)
```
